```python

from sklearn.datasets import load_wine
from sklearn.cluster import KMeans
import pandas as pd

# Load the wine dataset
wine = load_wine(as_frame=True)
print(wine.keys())  # Show the keys
```

```python

# Convert to DataFrame
df = wine.frame.copy()
```

```python

df.head()
```

```python

# Drop the target column
df = df.drop(columns=['target'])

from sklearn.preprocessing import StandardScaler
# Scale the features for better K-Means performance
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df)
df.head()
```

```python

print(scaled_features[:5])
```

```python

# Apply K-Means with n=3
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df['cluster'] = kmeans.fit_predict(scaled_features)
df.head()
```

```python

import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=df['cluster'], cmap='viridis', edgecolor='k', s=100)
plt.xlabel(wine.feature_names[0])
plt.ylabel(wine.feature_names[1])
plt.title('K-Means Clustering of Wine Data (n=3)')
plt.show()
```

```python

# Get user input for wine features
print("Enter the wine measurements:")
user_data = []
for feature in wine.feature_names:
    val = float(input(f"{feature}: "))
    user_data.append(val)

# Prepare input for prediction
user_data_scaled = scaler.transform([user_data])

# Predict cluster
cluster = kmeans.predict(user_data_scaled)[0]

print(f"The predicted cluster for the given wine is: {cluster}")
```

```python

import numpy as np

# Load again with target
wine = load_wine(as_frame=True)
df_full = wine.frame.copy()
X = df_full.drop(columns=['target'])
y = df_full['target']
feature_names = wine.feature_names

# Get class names and counts
class_names = wine.target_names
class_counts = np.bincount(y)

# Bar plot
plt.figure(figsize=(6, 4))
plt.bar(class_names, class_counts, color='skyblue', edgecolor='black')
plt.xlabel("Wine Class")
plt.ylabel("Count")
plt.title("Wine Dataset Class Distribution")
plt.show()

# Plot K-Means Clusters
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=kmeans.labels_, cmap='viridis', s=80, edgecolor='k')
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[1])
plt.title('K-Means Clustering (n=3)')

plt.subplot(1, 2, 2)
for label in range(len(class_names)):
    plt.scatter(
        X.loc[y == label, feature_names[0]],
        X.loc[y == label, feature_names[1]],
        label=class_names[label],
        s=100,
        edgecolor='k'
    )
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[1])
plt.title('Wine Dataset Scatter Plot')
plt.legend()
plt.tight_layout()
plt.show()
```
