# Unsupervised Learning on Iris Dataset

```python
from sklearn.datasets import load_iris
import pandas as pd

# Load Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
iris_df.head()
```

## Data Preprocessing

```python

from sklearn.preprocessing import StandardScaler

X = iris_df.drop('target', axis=1)

# Standardizing the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pd.DataFrame(X_scaled, columns=X.columns).head()
```

## Applying KMeans Clustering

```python

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Choosing the number of clusters using the Elbow Method
inertia = []
for i in range(1, 11):
    km = KMeans(n_clusters=i, random_state=42)
    km.fit(X_scaled)
    inertia.append(km.inertia_)

plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python

# Fit KMeans with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
iris_df['Cluster'] = kmeans.fit_predict(X_scaled)

iris_df.head()
```

## Visualizing Clusters

```python

import seaborn as sns

plt.figure(figsize=(8,6))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='Cluster', data=iris_df, palette='viridis')
plt.title('KMeans Clustering on Iris Data')
plt.show()
```

## PCA for Dimensionality Reduction

```python

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

iris_pca_df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
iris_pca_df['Cluster'] = iris_df['Cluster']

plt.figure(figsize=(8,6))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', data=iris_pca_df, palette='viridis')
plt.title('PCA - Iris Data Clusters')
plt.show()
```
