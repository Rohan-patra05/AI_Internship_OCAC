# Human Sleep Dataset – Data Visualization

This notebook walks through a handful of basic exploratory charts using **matplotlib** and **seaborn**.  
Dataset file: `expanded_human_sleep_dataset_augmented.csv`

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
sns.set_theme(style='whitegrid')
```

```python
df = pd.read_csv('expanded_human_sleep_dataset_augmented.csv')
print(df.head())
print('Shape:', df.shape)
```

## Basic Data Inspection

```python
df.describe().T
```

## 1️⃣ Matplotlib Visualizations

```python
# Histogram of total sleep
plt.figure(figsize=(6,4))
plt.hist(df['total_sleep'].dropna(), bins=20, color='steelblue', edgecolor='black')
plt.xlabel('Total Sleep (hours)')
plt.ylabel('Count')
plt.title('Distribution of Total Sleep Hours')
plt.show()
```

```python
# Pie chart gender
plt.figure(figsize=(5,5))
labels = df['gender'].value_counts().index
data = df['gender'].value_counts().values
plt.pie(data, labels=labels, autopct='%1.1f%%', startangle=140, colors=['#66b3ff','#ff9999'])
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()
```

```python
# Scatter body weight vs total sleep
plt.figure(figsize=(6,4))
plt.scatter(df['body_weight'], df['total_sleep'], alpha=0.7, color='purple')
plt.xlabel('Body Weight (kg)')
plt.ylabel('Total Sleep (hours)')
plt.title('Body Weight vs Total Sleep')
plt.show()
```

```python
# Bar average sleep per gender
avg_sleep_gender = df.groupby('gender')['total_sleep'].mean()
plt.figure(figsize=(6,4))
plt.bar(avg_sleep_gender.index, avg_sleep_gender.values, color=['#ff9999','#66b3ff'])
plt.ylabel('Average Total Sleep (hours)')
plt.title('Average Total Sleep by Gender')
plt.show()
```

```python
# Line plot age vs sleep
age_sorted = df.sort_values('age')
plt.figure(figsize=(7,4))
plt.plot(age_sorted['age'], age_sorted['total_sleep'], marker='o', linestyle='-', color='green')
plt.xlabel('Age (years)')
plt.ylabel('Total Sleep (hours)')
plt.title('Age vs Total Sleep')
plt.show()
```

```python
# Histogram BMI
plt.figure(figsize=(6,4))
plt.hist(df['bmi'].dropna(), bins=20, color='orange', edgecolor='black')
plt.xlabel('BMI')
plt.ylabel('Count')
plt.title('Distribution of BMI')
plt.show()
```

```python
# Boxplot sleep by gender
plt.figure(figsize=(6,4))
plt.boxplot([df[df['gender']=='Male']['total_sleep'].dropna(), 
             df[df['gender']=='Female']['total_sleep'].dropna()],
            tick_labels=['Male', 'Female'], patch_artist=True)
plt.ylabel('Total Sleep (hours)')
plt.title('Total Sleep by Gender')
plt.show()
```

## 2️⃣ Seaborn Visualizations

```python
# Histogram stress level
plt.figure(figsize=(6,4))
sns.histplot(data=df, x='stress_level', bins=10, kde=True, color='teal')
plt.title('Stress Level Distribution')
plt.show()
```

```python
# Regression screen time vs sleep
plt.figure(figsize=(6,4))
sns.regplot(data=df, x='screen_time_before_bed', y='total_sleep', scatter_kws={'alpha':0.6})
plt.title('Screen Time vs Total Sleep')
plt.show()
```

```python
# Boxplot exercise frequency vs sleep
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='exercise_frequency', y='total_sleep', hue='exercise_frequency', palette='pastel', legend=False)
plt.title('Exercise Frequency vs Total Sleep')
plt.show()
```

```python
# Violin gender vs BMI
plt.figure(figsize=(6,4))
sns.violinplot(data=df, x='gender', y='bmi', hue='gender', palette='muted', legend=False)
plt.title('BMI Distribution by Gender')
plt.show()
```

```python
# Heatmap correlation
plt.figure(figsize=(10,8))
corr = df.select_dtypes(include=[float,int]).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Pairplot selected vars
sns.pairplot(df[['total_sleep','age','bmi','stress_level','screen_time_before_bed']].dropna())
plt.show()
```

```python
# Countplot danger index
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='danger_index', hue='danger_index', palette='Set2', legend=False)
plt.title('Count of Danger Index')
plt.show()
```
