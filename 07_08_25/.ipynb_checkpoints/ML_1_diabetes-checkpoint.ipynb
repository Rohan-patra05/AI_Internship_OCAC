**Diabetes DATA SET**

The Diabetes dataset was used in R.A. Fisher's classic 1936 paper, The Use of Multiple Measurements in Taxonomic Problems, and can also be found on the UCI Machine Learning Repository.

It includes three diabetes species with 50 samples each as well as some properties about each flower. One flower species is linearly separable from the other two, but the other two are not linearly separable from each other.

The columns in this dataset are:

SepalLengthCm
SepalWidthCm
PetalLengthCm
PetalWidthCm
Target

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_diabetes

# Load the diabetes dataset
diabetes = load_diabetes(as_frame=True)
df = diabetes.frame
df['target'] = diabetes.target

# Group by target and compute mean of each feature
mean_df = df.groupby('target').mean()

# Plotting
plt.figure(figsize=(8, 6))

for column in diabetes.feature_names:
    plt.plot(mean_df.index, mean_df[column], marker='o', label=column)

plt.xlabel('Target Variable (Disease Progression)')
plt.ylabel('Mean Feature Value')
plt.title('Mean Feature Values per Diabetes Target Class')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_diabetes

# Load the diabetes dataset
diabetes = load_diabetes(as_frame=True)
df = diabetes.frame
df['target'] = diabetes.target

# Group by target and compute mean of each feature
mean_df = df.groupby('target').mean()

# Create the bar plot
mean_df.plot(kind='bar', figsize=(10, 6))

plt.xlabel('Target Variable (Disease Progression)')
plt.ylabel('Mean Feature Value')
plt.title('Mean Feature Values per Diabetes Target Class')
plt.xticks(rotation=0)  # Rotate x-axis labels if needed
plt.legend(title='Features')
plt.grid(axis='y') # Add grid lines only on the y-axis
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X = diabetes.data  # Features
y = diabetes.target # Target variable

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# You can now see the shapes of the resulting arrays
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Fit on training data
X_train_scaled = scaler.fit_transform(X_train)

# Transform test data
X_test_scaled = scaler.transform(X_test)

# Optional: Check the mean and std dev after scaling
print(f"Mean of scaled features (train): {X_train_scaled.mean(axis=0)}")
print(f"Std dev of scaled features (train): {X_train_scaled.std(axis=0)}")
```

```python
import pandas as pd

# Load dataset
diabetes = load_diabetes()
X = diabetes.data
y = diabetes.target
feature_names = diabetes.feature_names

# Scale data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

# Convert to DataFrame
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=feature_names)

# Display
print(X_train_scaled_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert to DataFrame for easy handling
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=feature_names)

# Select first sample
sample_index = 0
sample_data = X_train_scaled_df.iloc[sample_index]

# Plot bar graph for first sample
plt.figure(figsize=(8, 5))
sample_data.plot(kind='bar', color='orange', edgecolor='black')
plt.title(f"Scaled Feature Values for Training Sample {sample_index}")
plt.xlabel("Features")
plt.ylabel("Scaled Value")
plt.axhline(0, color='gray', linestyle='--')
plt.grid(axis='y', linestyle=':', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Import necessary libraries
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# 1️⃣ Train a Logistic Regression model
model = LogisticRegression()
model.fit(X_train_scaled, y_train)

# 2️⃣ Predict on test data
y_pred = model.predict(X_test_scaled)
```

```python
# 3️⃣ Evaluate model performance
print("✅ Accuracy:", accuracy_score(y_test, y_pred))
print("\n✅ Classification Report:\n", classification_report(y_test, y_pred))
```

```python
# Convert to binary classification: e.g., target > 140 is "high risk"
y_test_class = (y_test > 140).astype(int)
y_pred_class = (y_pred > 140).astype(int)

# Confusion matrix
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test_class, y_pred_class)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Low', 'High'],
            yticklabels=['Low', 'High'])
plt.title("Confusion Matrix")
plt.xlabel("Predicted Class")
plt.ylabel("Actual Class")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
print("R² Score:", r2_score(y_test, y_pred))
```

```python
import joblib

# Save the trained model
joblib.dump(model, 'logistic_regression_model.pkl')

# Save the fitted scaler
joblib.dump(scaler, 'diabetes.pkl')

print("Model and scaler saved successfully!")
```

```python
import joblib
import numpy as np
import pandas as pd

# Load the saved model and scaler
loaded_model = joblib.load('diabetes.pkl')    # should be a regressor
loaded_scaler = joblib.load('diabetes.pkl')  # StandardScaler

# Feature order for diabetes dataset
feature_names = ['age', 'sex', 'bmi', 'bp', 's1', 's2', 's3', 's4', 's5', 's6']

# Get 10 inputs from user
user_inputs = []
for feature in feature_names:
    val = float(input(f"Enter value for {feature}: "))
    user_inputs.append(val)

# Create DataFrame with feature names
custom_input_df = pd.DataFrame([user_inputs], columns=feature_names)

# Scale the input
custom_input_scaled = loaded_scaler.transform(custom_input_df)

# Predict the disease progression score
predicted_value = loaded_model.predict(custom_input_scaled)

print("\nPredicted disease progression score:", predicted_value[0])
```
